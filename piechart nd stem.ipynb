# pie chart plot

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x,labels=y)


plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c)


plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["y","g","b","r"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,labeldistance=1,startangle=0,textprops={"fontsize":15})


plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["y","g","b","r"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,
        radius=0.8,labeldistance=1,startangle=0,textprops={"fontsize":15},counterclock=False)


plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["y","g","b","r"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,
        radius=0.8,labeldistance=1,startangle=0,textprops={"fontsize":15},
        counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"},rotatelabels=True)


plt.title("Wscube Tech")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["y","g","b","r"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,
        radius=0.8,labeldistance=1,startangle=0,textprops={"fontsize":15},
        counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"},rotatelabels=True)


plt.title("Wscube Tech")
plt.legend(loc=2)
plt.show()
```

# stem plot 


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]

plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=9)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[6,5,4,3,2,1]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python",use_line_collection=False,orientation="horizontal")
plt.show()
```
