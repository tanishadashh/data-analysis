# barplot


```python
import matplotlib.pyplot as plt 
x=["python","c","java","c++"]
y=[85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["python","c","java","c++"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=10)
plt.ylabel("no")
plt.title("wscube")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["python","c","java","c++"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=10)
plt.ylabel("no")
plt.title("wscube")
plt.bar(x,y,width=0.4,color="y")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["python","c","java","c++"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=10)
plt.ylabel("no")
plt.title("wscube")
c=["r","b","g","m"]
plt.bar(x,y,width=0.4,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["python","c","java","c++"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=10)
plt.ylabel("no")
plt.title("wscube")
width=0.2
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,z,width=0.4,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x=["python","c","java","c++"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("no")
plt.title("wscube")
plt.bar(p,y,width=0.4,color="r",label="popularity")
plt.bar(p1,z,width=0.4,color="y",label="popularity1")

plt.xticks(p+width,x)
plt.legend()
plt.show()


```

# scatterplot

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatterplot",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatterplot",fontsize=20)
plt.xlabel("day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
plt.scatter(Day,no,c=colors)
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[200,500,100,300,200,400,600]
plt.scatter(Day,no,c=colors,s=sizes,alpha=0.5)
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[200,500,100,300,200,400,600]
plt.scatter(Day,no,c=colors,s=sizes,alpha=0.5,marker="*")
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[200,500,100,300,200,400,600]
plt.scatter(Day,no,c=colors,s=sizes,cmap="viridis")
plt.colorbar()
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[200,500,100,300,200,400,600]
plt.scatter(Day,no,c=colors,s=sizes,cmap="plasma",alpha=0.2)
t=plt.colorbar()
t.set_label("colorbar")
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt 
Day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,2,6]
colors=[10,49,30,29,56,20,30]
sizes=[200,500,100,300,200,400,600]
plt.scatter(Day,no,c=colors,s=sizes,cmap="plasma",alpha=0.2)
plt.scatter(Day,no2,color="r",s=sizes,alpha=0.3)
t=plt.colorbar()
t.set_label("colorbar")
plt.title("Scatterplot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```
