```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
y=[3,2,4,5,6]

plt.step(x,y,marker="o",color="g")
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
y=[3,2,4,5,6]

plt.step(x,y,marker="o",color="g",ms=10,mfc="g")
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
y=[3,2,4,5,6]

plt.step(x,y,marker="o",color="g",ms=10,mfc="g")
plt.grid()

plt.legend(loc=2)
plt.show()
```
